```python
# Importing pandas
import pandas as pd
import matplotlib.pyplot as plt

# SQL Alchemy
from sqlalchemy import create_engine

# Defining path to database
engine = create_engine("postgres://postgres:1234@localhost:5432/employees")
conn = engine.connect()
```

```python
# Creating dataframes of the relevent tables
employees_data = pd.read_sql("SELECT * FROM employees", conn)
salary_data = pd.read_sql("SELECT * FROM salaries", conn)
title_data = pd.read_sql("SELECT * FROM titles", conn)

# Keeping the last title held by employee for data consistency
title_data_last = title_data.drop_duplicates(["emp_no"], keep = "last")
```

```python
# Merging employees and salaries dataframes and choosing to do inner join to remove null values, if any.
salary_emp_merged = pd.merge(employees_data, salary_data, how = "inner", on = "emp_no")
salary_emp_merged
```

```python
# Merging merged dataframe with salaries to get titles of employees
sal_emp_title_merged = pd.merge(salary_emp_merged, title_data_last, how = "left", on = "emp_no")

# Selecting only the relevent columns 
sal_title = sal_emp_title_merged[["title", "salary"]]
sal_title
```

```python
# Converting salary data to float
sal_title["salary"] = sal_title["salary"].astype(float)
sal_title.head()
```

```python
# Create a histogram to visualize the most common salary ranges for employees.
sal_title.plot.hist(by = "salary")
plt.title("Most Common Salary Range")
plt.xlabel("Salary USD")

# Printing results
plt.show()
```

```python
# Grouping by salary to get average salary per title
sal_group = sal_title.groupby("title")
sal_group_mean = sal_group.mean()
```

```python
# Create a bar chart of average salary by title.
sal_group_mean.plot.bar()
plt.title("Average Salary by Title")
plt.xlabel("Job Title")

# Printing results
plt.show()
```
